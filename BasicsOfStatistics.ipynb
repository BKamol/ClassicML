- **Генеральная совокупность** - множество всех исследуемых объектов
- **Выборка** - подмножество генеральной совокупности, репрезентативное, чтобы можно было интерполировать выводу на генеральную совокупность.

- **Простая случайная выборка** - объекты в выборке отобраны случайным образом.
- **Стратифицированная выборка** - совокупность разбивается на страты(группы), затем из каждой страты случайным образом отбираются объекты.
- **Групповая выборка** - из многиз кластеров генеральной совокупности выбираются некоторые, а объекты случайным образом отбираются только из них.

Изучаемая переменная может быть **количественной**, например, рост и количество детей, а может быть **качественной**, например пол. Причем количественные также могут быть **непрерывными** или **дискретными**, а качественные - **номинативными**(не допускают сравнение) и **ранговыми**(допускают сравнение).

### Меры центральной тенденции

- **Мода** - самое частое значение измеряемого признака.
- **Медиана** - значение признака, которое делит упорядоченное множество признаков пополам
- **Среднее значение** - Сумма признаков, деленное на их число

- $M[X+C] = M[X] + C$, $M[XC] = CM[x]$

### Меры изменчивости

- **Размах** - разность максимального и минимального значений признака.
- **Дисперсия** - среднее квадратов отклонений значений признаков от среднего значения выборки.
- **Среднеквадратическое отклонение** - корень из дисперсии.
- $D[X+C]$ = $D[X]$, $D[XC]$ = $C^2D[X]$
- $d[X+C] = d[X], d[XC] = Cd[X]$

### Квартили распределения

- **Квартили** - три точки(значения признаков), делящие данные на 4 равные части

### Нормальное распределение

- Унимодально
- Симметрично
- Отклонения наблюдений от среднего подчиняются определенному вероятностному закону.

- Стандартизация или z-преобразование - это преобразование распрделения X в Z, при котором $E[Z]=0, D[Z]=1$, самj преобразование 
- $Z_i = (X_i - E[X])/(std) ,      std=sqrt(D[X])$

По правилу "трех" сигм
- $E[X] +- std =(примерно)  68$ % наблюдений
- $E[X] +- 2*std =(примерно) 95$ % наблюдений
- $E[X] +- 3*std =(примерно) 100$ % наблюдений

### Центральная предельная теорема

Если увеличить размер выборки, то сильные отклонения выборочных средних от истинного среднего будут возникать реже.
Пусть есть признак, распределенный КАК  УГОДНО* с некоторым средним и некоторым стандартным отклонением. Тогда, если мы будем выбирать из этой совокупности выборки объема n, то их средние тоже будут распределены нормально со средним равным среднему признака в ГС и отклонением стандартным отклонением, se, формулу которого мы уже приводили в прошлых шагах.

### Доверительный интервал

**Доверительный интервал** - это интервал $E[X]+-1.96se$. В 95% случаев доверительынй интервал содержит среднее генеральной совокупности.
