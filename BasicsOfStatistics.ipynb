- **Генеральная совокупность** - множество всех исследуемых объектов
- **Выборка** - подмножество генеральной совокупности, репрезентативное, чтобы можно было интерполировать выводу на генеральную совокупность.

- **Простая случайная выборка** - объекты в выборке отобраны случайным образом.
- **Стратифицированная выборка** - совокупность разбивается на страты(группы), затем из каждой страты случайным образом отбираются объекты.
- **Групповая выборка** - из многиз кластеров генеральной совокупности выбираются некоторые, а объекты случайным образом отбираются только из них.

Изучаемая переменная может быть **количественной**, например, рост и количество детей, а может быть **качественной**, например пол. Причем количественные также могут быть **непрерывными** или **дискретными**, а качественные - **номинативными**(не допускают сравнение) и **ранговыми**(допускают сравнение).
